```python
# مدخلات
ai_prediction = "مريض"    # أو "سليم"
nds_score = 7
nss_score = 4

# تحويل لثنائية
ai_binary = 1 if ai_prediction == "مريض" else 0
nds_binary = 1 if nds_score >= 6 else 0
nss_binary = 1 if nss_score >= 5 else 0

# أوزان مقترحة
weight_nds = 1.2
weight_ai  = 1.0
weight_nss = 0.9

# حساب الدرجة
score = ai_binary*weight_ai + nds_binary*weight_nds + nss_binary*weight_nss
threshold = (weight_ai + weight_nds + weight_nss) / 2

if score >= threshold:
    final_decision = "PDN مؤكد"
else:
    final_decision = "سليم على الأغلب"

print(final_decision)
```
