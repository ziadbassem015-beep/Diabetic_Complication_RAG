```python
def diabetic_foot_ulcer_screening():
    print("=== Diabetic Foot Ulcer Screening ===")

    wound_not_healing = input("Do you have wounds that do not heal? (yes/no): ").lower()
    numbness = input("Do you feel numbness or loss of sensation? (yes/no): ").lower()
    pain = input("Is there pain in the foot? (yes/no): ").lower()
    color_change = input("Any change in skin color? (yes/no): ").lower()
    swelling = input("Any swelling or redness? (yes/no): ").lower()
    discharge = input("Any discharge or bad odor? (yes/no): ").lower()
    tight_shoes = input("Do you wear tight shoes? (yes/no): ").lower()

    score = 0

    if wound_not_healing == "yes":
        score += 3

    if numbness == "yes":
        score += 2

    if pain == "no" and wound_not_healing == "yes":
        score += 2

    if color_change == "yes":
        score += 2

    if swelling == "yes":
        score += 2

    if discharge == "yes":
        score += 3

    if tight_shoes == "yes":
        score += 1

    print("\n=== Result ===")

    if score >= 8:
        print(" High Risk: Possible diabetic foot ulcer - seek medical attention immediately!")
    elif score >= 4:
        print(" Moderate Risk: Monitor closely and consult a doctor.")
    else:
        print("Low Risk: Maintain foot care and monitor regularly.")

    print(f"Risk Score: {score}")


diabetic_foot_ulcer_screening()
```
