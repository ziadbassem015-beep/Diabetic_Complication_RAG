```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# ========================
# 1️⃣ محرك النصائح المخصصة (Personalized Advice Engine)
# ========================
def get_custom_tips(bleeding, redness, swelling, calculus, breath, smoking):
    tips = []

    if bleeding >= 2:
        tips.append("🩸 <b>للنزيف:</b> لا تتوقف عن التفريش بسبب الدم؛ النزيف دليل التهاب بكتيري يحتاج تنظيفاً أعمق ولكن بلطف.")

    if redness >= 1:
        tips.append("🎨 <b>للون:</b> احمرار اللثة يعني احتقان الأوعية؛ استخدم مضمضة تحتوي على مياه دافئة وملح لتهدئة الأنسجة.")

    if swelling >= 1:
        tips.append("🎈 <b>للتورم:</b> تجنب الأطعمة الصلبة التي قد تجرح اللثة المنفوخة، واستخدم فرشاة أسنان فائقة النعومة (Ultra-Soft).")

    if calculus >= 1:
        tips.append("🧱 <b>للجير:</b> الجير لا يزول بالفرشاة؛ أنت بحاجة لتنظيف احترافي في العيادة لمنع تآكل العظم السنخي.")

    if breath == 1:
        tips.append("🌬️ <b>للرائحة:</b> اهتم بتنظيف سطح اللسان، فهو مخزن رئيسي للبكتيريا المسببة للرائحة.")

    if smoking == 1:
        tips.append("🚬 <b>للتدخين:</b> النيكوتين يضيق الأوعية ويخفي النزيف؛ لثتك قد تكون أسوأ مما تبدو عليه، لذا الفحص الدوري ضروري.")

    if not tips:
        tips.append("✨ لثتك تبدو مثالية! استمر على غسل الأسنان مرتين يومياً واستخدام الخيط.")

    return tips

def calculate_gum_health(bleeding, redness, swelling, calculus, breath, smoking):
    weights = {'bleeding': 3, 'redness': 2, 'swelling': 2, 'calculus': 2, 'breath': 1, 'smoking': 1}
    total_score = (bleeding * weights['bleeding']) + (redness * weights['redness']) + \
                  (swelling * weights['swelling']) + (calculus * weights['calculus']) + \
                  (breath * weights['breath']) + (smoking * weights['smoking'])

    if total_score <= 5:
        status, c1, c2, icon = "لثة سليمة", "#11998e", "#38ef7d", "✅"
    elif total_score <= 11:
        status, c1, c2, icon = "التهاب خفيف", "#f7971e", "#ffd200", "⚠️"
    elif total_score <= 18:
        status, c1, c2, icon = "التهاب متوسط", "#cb2d3e", "#ef473a", "🚨"
    else:
        status, c1, c2, icon = "التهاب شديد", "#4b1248", "#f03023", "🛑"

    return total_score, status, c1, c2, icon

# ========================
# 2️⃣ واجهة المستخدم (Widgets)
# ========================
style = {'description_width': 'initial'}
redness_w = widgets.RadioButtons(options=[('🩷 وردي طبيعي', 0), ('❤️ أحمر زاهي', 1), ('💜 داكن', 2)], description='<b>اللون:</b>', style=style)
bleeding_w = widgets.SelectionSlider(options=[('لا يوجد', 0), ('بسيط', 1), ('واضح', 2), ('تلقائي', 3)], description='<b>النزيف:</b>', style=style)
swelling_w = widgets.Dropdown(options=[('لا يوجد', 0), ('بسيط', 1), ('شديد', 2)], description='<b>التورم:</b>', style=style)
calculus_w = widgets.Dropdown(options=[('لا يوجد', 0), ('بسيط', 1), ('كثيف', 2)], description='<b>الجير:</b>', style=style)
breath_toggle = widgets.ToggleButton(value=False, description='رائحة فم؟', icon='check', button_style='info')
smoke_toggle = widgets.ToggleButton(value=False, description='مدخن؟', icon='smoking', button_style='info')
analyze_button = widgets.Button(description="تحليل النتائج وعرض النصائح", button_style='success', layout={'height': '50px', 'width': '300px', 'margin': '20px auto'})
result_output = widgets.Output()

# ========================
# 3️⃣ دالة التحليل والعرض
# ========================
def analyze_values(b):
    with result_output:
        clear_output()
        score, label, c1, c2, icon = calculate_gum_health(bleeding_w.value, redness_w.value, swelling_w.value, calculus_w.value, 1 if breath_toggle.value else 0, 1 if smoke_toggle.value else 0)
        tips = get_custom_tips(bleeding_w.value, redness_w.value, swelling_w.value, calculus_w.value, 1 if breath_toggle.value else 0, 1 if smoke_toggle.value else 0)

        tips_html = "".join([f"<li style='margin-bottom:8px;'>{tip}</li>" for tip in tips])

        display(HTML(f"""
            <div style="background: linear-gradient(135deg, {c1} 0%, {c2} 100%); padding: 25px; border-radius: 20px; color: white; font-family: sans-serif; text-align: right; direction: rtl; box-shadow: 0 10px 25px rgba(0,0,0,0.2); max-width: 500px; margin: auto;">
                <div style="font-size: 40px; text-align: center;">{icon}</div>
                <h2 style="text-align: center; margin: 5px 0;">{label} ({score}/23)</h2>
                <hr style="border: 0; border-top: 1px solid rgba(255,255,255,0.3); margin: 15px 0;">
                <h4 style="margin-bottom: 10px;">💡 نصائح مخصصة لحالتك:</h4>
                <ul style="list-style-type: none; padding-right: 10px; font-size: 15px; line-height: 1.5;">
                    {tips_html}
                </ul>
            </div>
        """))

analyze_button.on_click(analyze_values)
display(widgets.VBox([widgets.HTML("<h2 style='text-align:center;'>🏥 طبيب اللثة الافتراضي</h2>"), redness_w, bleeding_w, swelling_w, calculus_w, widgets.HBox([breath_toggle, smoke_toggle], layout={'justify_content':'center'}), analyze_button], layout={'padding':'20px', 'background':'white', 'border-radius':'15px', 'width':'500px', 'margin':'auto'}), result_output)
```
